# Basic Line Ploting

```python
#import req functions
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,10) # 10sample points range from 0-5
print(x)
print(type(x))
y=x**2 # EQUATION y=x^2
print(y)
```

```python
plt.figure() # << It is canvas 
# plt.plot(x,y,'r',ls='--')   # r = red color, ls='--' << line space   lw << Line Width
plt.plot(x,y,'b-.',lw=2)  #'r-*'
plt.xlabel("Time Stamp")
plt.ylabel("Temperature value")
plt.title("My first graph", fontsize=10)
plt.grid()
plt.show() # mandatory without this it wont show uotput
```

```python
plt.plot(x, y, 'r-*',label='line1',lw=0.2 )
plt.plot(y, x, 'b--' , label='line2')
plt.xlabel('timesample')
plt.ylabel('temperature values')
plt.title('my second graph')
plt.grid() 
# default on top right corner
plt.legend(loc=2) #identification of lines lables and there colors(NOTE : label sgoud be defined)

plt.show()
```

```python
list1=[1,3,7,5,10]
list2=[1,3,4,7,8]

plt.figure()
plt.plot(list1,list2,'b*-', label='rateofprice',lw=1.0,ms=5.0)
plt.xlabel("timestamp")
plt.ylabel("rupee value")
plt.title("Rupee Currency rate")
plt.legend()
plt.grid()
plt.show()
```

## Subploting

2 graphs in graph


```python
plt.subplot(1,2,1) # specify howmany columns >> 1strow 2columns and it is 1st one 
plt.plot(x,y,'r--') # y wrt x and r-- dashed lines in red 
plt.subplot(1,2,2)   # 1st row 2 columns and it is second one
plt.plot(y,x,'b-*')# x wrt y and b-- dashed* lines in red
plt.show()
```

```python
#FOUR GRAPHS 2rows and 2 columns

plt.subplot(2,1,1) #  2rows 1column and it is "1st one" 
plt.plot(x,y,'r--') # y wrt x and r-- dashed lines in red 
plt.subplot(2,1,2)   # 2rows 1column and it is "second one"
plt.plot(y,x,'b-*')# x wrt y and b-- dashed* lines in red
plt.show()
```

```python
#FOUR GRAPHS 2rows and 2 columns

plt.subplot(2,2,1) #  two-rows 2columns and it is "1st one" 
plt.plot(x,y,'m+') # y wrt x and r-- dashed lines in red 

plt.subplot(2,2,2)   # 2rows 2columns and it is "second one"
plt.plot(y,x,'b-*')# x wrt y and b-- dashed* lines in red

plt.subplot(2,2,3) #2rows 2columns  and it is "3st one" 
plt.plot(x**2,y,'g:') # y wrt x and r-- dashed lines in red 

plt.subplot(2,2,4)   # 2rows 2columns  and it is "FOURTH one"
plt.plot(y*2,x**2,'y--')# x wrt y and b-- dashed* lines in red

plt.show()
```

# Bar and Scatter Plots

```python
list1=[1,3,7,5,11]
list2=[1,3,4,7,8]
#x and y shpuld be of same length *
plt.bar(list1,list2,color='black')
plt.grid()
```

```python
# Horizantal BARGRAPH  (barh)

list1=[1,3,7,5,11]
list2=[1,3,4,7,8]
#x and y shpuld be of same length *
plt.barh(list1,list2,color='black')  
plt.grid()
```

```python
list1=[1,3,8,12,11]
list2=[2,3,3,8,8]

l3=[4,5,6,8,6]
l4=[5,6,4,7,6]

#x and y shpuld be of same length *
plt.bar(list1,list2,color='b')
plt.bar(l3,l4,color='green')

plt.grid()
```

```python

# the positions of the bars on the x-axis
x=[1,2,3,4,5]

#the y axis points
y1=[40,40,20,12,33]
y2=[12,35,3,10,25]

plt.bar(x,y1,color='black')
plt.bar(x,y2,color='m')

plt.show()
```

```python

# y2 on the top of the Y1 by bottom attribute
x=[1,2,3,4,5]

#the y axis points
y1=[40,40,20,12,33]
y2=[12,35,3,10,25]

plt.bar(x,y1,color='black')
plt.bar(x,y2,color='skyblue',bottom=y1)

plt.show()
```

```python
# Naming x-axis CUSTOM

x=[1,2,3,4,5]

#the y axis points
y1=[40,40,20,12,33]
y2=[12,35,3,10,25]

#CUSTOM X- AXIS
names=['A','B','C','D','E']  #LENGTH OF X and names List should be same
plt.xticks(x, names, fontweight='bold')

plt.xlabel("Group")

plt.bar(x,y1,color='black')
plt.bar(x,y2,color='yellow',bottom=y1)
# plt.bar(x,y2,color='m')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

## Problem 1
plot sin, cos, and sin+cos waves in different figures

```python
x=np.arange(0,20,0.1)
    
plt.figure(figsize=(15, 7))

plt.subplot(2,2,1)
plt.plot(x,np.sin(x),'b',lw=1)
plt.title("sin wave")
plt.grid()

plt.subplot(2,2,2)
plt.plot(x,np.cos(x),'g-',lw=1)
plt.title("cos wave")
plt.grid()

plt.subplot(2,2,3)
plt.plot(x,np.sin(x)+np.cos(x),'r-',lw=1)
plt.title("sin+cos wave")
plt.grid()

plt.subplot(2,2,4)
plt.plot(x,np.sin(x),'g-',lw=1)
plt.plot(x,np.cos(x),'b-',lw=0.6)
plt.plot(x,np.tan(x),'r-.',lw=1)
plt.title("sin+cos+tan wave")
plt.grid()

plt.show()
```

## Problem 2

```python
x=np.arange(0,20,0.1)
sin1=[]
sin2=[]
sin3=[]


for point in x:
    sin1.append(math.sin(point-1))

for point in x:
    sin2.append(math.sin(point))
    
for point in x:
    sin3.append(math.sin(point+1))

plt.figure(figsize=(20, 5))
plt.plot(x,sin1,'b-',lw=2)
plt.plot(x,sin2,'r-',lw=2)
plt.plot(x,sin3,'g-',lw=1)
plt.grid() 
plt.show()

```

```python
x=np.arange(0,20,0.1)

#finding the 120 degrees
pi=math.pi
deg120=pi/3+pi/3

sin1=[math.sin(point-deg120) for point in x]
sin2=[math.sin(point) for point in x]
sin3=[math.sin(point+deg120) for point in x]

plt.figure(figsize=(20, 5))
plt.plot(x,sin1,'b-',lw=2)
plt.plot(x,sin2,'r-',lw=2)
plt.plot(x,sin3,'g-',lw=1)
plt.grid() 
plt.show()
```
