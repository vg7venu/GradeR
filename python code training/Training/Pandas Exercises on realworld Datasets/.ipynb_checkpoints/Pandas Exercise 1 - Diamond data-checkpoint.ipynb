# Pandas Exercise 1 - Diamond data

The following csv data file contains data of diamonds with following characteristics.

carat - diamond's apparent size <br>
cut - how good the diamond shaping is <br>
color - colour of a diamond as letter representation <br>
clarity - clarity of diamond <br>
depth - depth of diamond <br>
table - top surface of diamond <br>
price - price of diamond <br>
x - x dimension <br>
y - y dimension<br>
z - z dimension <br>

Read the file and find the following:
1. Drop the "unnamed" column first
2. Number of diamonds of each colour
3. Number of diamonds of each type of clarity
4. Number of diamonds with carat value greater than or equal to 0.25
5. Maximum depth value of each pair of color and clarity
6. Maximum price of each clarity
7. Least depth of each color
8. Number of diamonds which have table value between 55 to 60
9. Highest price of premium cut of each color
10. Create a new column named 'volume' and the formula is volume=x*y*z
11. Find the number of diamonds with maximum price/volume of each colour.

Also plot the following using pandas plot function:
1. Bar graph of color
2. Pie plot of clarity
3. Histograms of Price and depths colums as two separate plots 
4. A scatter plot of carat vs Price, volume vs Price, depth vs Price in a subplot of 1x3.

```python
import pandas as pd
df = pd.read_csv("diamonds.csv") #Comma seperated files :: carat,cut,color
# df = pd.read_excel("diamonds.csv") << To read excel file
df.head()
```

## 1 Drop the "unnamed" column first

```python
df = df.drop('Unnamed: 0',axis=1,)
```

```python
df.head()
```

## 2. Number of diamonds of each colour

```python
df['color'].value_counts()
```

## 3.Number of diamonds of each type of clarity

```python
df['clarity'].value_counts()
```

```python
df.head()
```

## 4. Number of diamonds with carat value greater than or equal to 0.25

```python
df['carat'][df['carat']>=0.25].count()
```

## 5. Maximum depth value of each pair of color and clarity

```python
df.groupby(['clarity','color'])['depth'].max()
# for i,x in df.groupby(['clarity','color']):
#     print(i,'\n',x)
```

```python
df.head()
```

## 6 . Maximum price of each clarity

```python
df.groupby('clarity')['price'].max()
```

## 7. Least depth of each color

```python
df.groupby('color')['depth'].min()
```

## 8. Number of diamonds which have table value between 55 to 60

```python
df.head()
```

```python
len(df[(df['table']>55.0) & (df['table']<60.0)].index)
```

## 9 Highest price of premium cut of each color

```python
df.head()
```

```python
df.groupby(['cut','color']).max().loc['Premium']['price']

#sir code :
# fill_data = df[df['cut']=='Premium']
# print(fill_data.groupby(['color'])[['cut','price']].max() )
```

## 10 Create a new column named 'volume' and the formula is volume=xyz

```python
df.head()
```

```python
df['volume'] = df['x']*df['y']*df['z']
df
```

## 11 Find the number of diamonds with maximum price/volume of each colour.

```python
df.groupby('color')[['price','volume']].max()
```

## BAR GRAPG COLOR

```python
# df.plot.bar(x=df.index, y=df['color'],)
df['color'].value_counts().plot(kind='bar')
```

## pie plot of clarity

```python
df['clarity'].value_counts().plot(kind='pie')
```

## Histograms of Price and depths colums as two separate plots

```python
df.head()
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.subplot(1,2,1)
df['price'].value_counts().plot(kind='hist')
plt.plot()



plt.subplot(1,2,2)
df['volume'].value_counts().plot(kind='hist')

plt.show()

```

```python
plt.figure(figsize=(30,13))
plt.subplot(1,3,1)
plt.scatter(df['carat'],df['price'],color='m',marker=1)
plt.subplot(1,3,2)
plt.scatter(df['volume'],df['price'],c='y')
plt.subplot(1,3,3)
plt.scatter(df['depth'],df['price'],c='r')
plt.show()
```
