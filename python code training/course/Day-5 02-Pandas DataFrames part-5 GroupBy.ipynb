# Pandas - GroupBy

> GroupyBy allows you to fgroup together rows based off of a column and perform an aggregate function on them.
<br>
###### Aggregatefunction: any function that takes many values and sipins or outputs out single value

```python
import pandas as pd
import numpy as np
```

```python
data = {
    'Company':['GOOG','GOOG','MSFT','MSFT','FB','FB'],
    'Person':['Sam','Charlie','Amy','Vanessa','Carl','Sarah'],
    'Sales':[200,120,340,124,243,350]
}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
byComp = df.groupby('Company')
```

```python
#Mean every value column
byComp.mean()
```

```python
byComp.std()
```

```python
byComp.sum()
```

```python
#WE GOT A DATAFRAME so we can do this
#INDEX Become Company Name
byComp.sum().loc['FB']
```

```python
df.groupby('Company').count()
```

```python
#python compare strings alphabetical order
df.groupby('Company').max()
```

```python
df.groupby('Company').min()
```

```python
#For sales it will give USEFUL INFO
#NOTE If we Got any other Value Column we can get that value too in seprate stuff
df.groupby('Company').describe()
```

```python
#Changing Format
df.groupby('Company').describe().transpose()
```
